# Linear Regression

```python
# Importar los módulos necesarios para el desarrollo un modelo de regresión lineal.

%matplotlib notebook
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
```

```python
#Importar la Database
df = pd.read_csv("casas.csv")
df.head(10)
```

```python
# Dar valor a las variables
x = df["N_Habitaciones"]
y = df["Valor_CLP"]
```

```python
# Calcule la regresión lineal de mínimos cuadrados para dos conjuntos de datos.
slope, intercept, r, p, std_err = stats.linregress(x, y)
```

```python
# Utilizar la función auxiliar que usa la pendiente e intersección 
#calculadas para devolver un nuevo valor.

def fitted_value(x):
    return slope * x + intercept
```

```python
# Calcular los valores ajustados para el eje ‘y’ usando una función auxiliar.
new_values = list(map(fitted_value, x))
```

```python
# Visualización del modelo.
plt.plot(x, y, 'o', label='original data')

# Draw the line or linear regression
plt.plot(x, new_values, 'r', label='fitted line')

plt.legend()
```
